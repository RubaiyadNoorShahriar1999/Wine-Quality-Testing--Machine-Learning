```python
features = ['citric acid','alcohol','residual sugar','pH','total sulfur dioxide']  #input values

x = wine_data[features]
y = wine_data['quality'] 
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()
y= label_encoder.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.10 , random_state=3,shuffle=True)
```

```python
s_c = StandardScaler()
X_train=s_c.fit_transform(X_train)
X_test=s_c.fit_transform(X_test)
```

```python
Knn = KNeighborsClassifier (n_neighbors=7,p=1)
Knn.fit(X_train,y_train)
```

```python
y_pred=Knn.predict(X_test)
y_pred
```

```python
y_test
```

```python
c_m = confusion_matrix(y_test,y_pred)
c_m
```

```python
print("Wrong values predicted: ", (y_test!=y_pred).sum())
print("Total values: ",((y_test==y_pred).sum()+(y_test!=y_pred).sum()))
```

```python
print('percentage Accuracy using KNN is : ',100*accuracy_score(y_test,y_pred), "%")
```
